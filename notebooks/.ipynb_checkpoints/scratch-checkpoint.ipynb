```python
data = utils.load_it(directory+'Teams_logs.pk')
```

```python
# Computing individual performance rates.
hardness_weights = {
    pogs_jeopardy_log_lib.Level.EASY: 1,
    pogs_jeopardy_log_lib.Level.MEDIUM: 2,
    pogs_jeopardy_log_lib.Level.HARD: 3}
minimum_number_of_questions_to_answer = 44

questions = data[37].game_info.questions
individual_performances = {}
for team_id, team_log in data.items():
    individual_performances[team_id] = defaultdict(
        lambda: {'#correct': 0,
                 '#questions': 0,
                 '#hardness_weighted_correct': 0,
                 '#hardness_weighted_questions': 0})
    for qid, team_answers in team_log.individual_answers_chosen.items():
        hardness_weight = hardness_weights[questions[qid].level]
        for member in team_log.members:
            if member in team_answers:
                individual_performances[team_id][member]['#questions'] += 1
                individual_performances[team_id][member]['#hardness_weighted_questions'] += hardness_weight
                if questions[qid].answer == team_answers[member]:
                    individual_performances[team_id][member]['#correct'] += 1
                    individual_performances[team_id][member]['#hardness_weighted_correct'] += hardness_weight
individual_performance_rates = defaultdict(dict)
for team_id, team_log in data.items():
    for member in team_log.members:
        if individual_performances[team_id][member]['#questions'] >= minimum_number_of_questions_to_answer:
            correct_rate = individual_performances[team_id][member]['#correct'] / individual_performances[team_id][member]['#questions']
            hardness_weighted_correct_rate = individual_performances[team_id][member]['#hardness_weighted_correct'] / individual_performances[team_id][member]['#hardness_weighted_questions']
            individual_performance_rates[team_id][member] = {
                'correct_rate': correct_rate,
                'hardness_weighted_correct_rate': hardness_weighted_correct_rate}
            
# Adding 2 more features to individual performance rates to compare one to other teammates.
for team_id, individual_performance in individual_performance_rates.items():
    team_correct_rate_mean = 0
    team_hardness_weighted_correct_rate_mean = 0
    for perf in individual_performance.values():
        team_correct_rate_mean += perf['correct_rate']
        team_hardness_weighted_correct_rate_mean += perf['hardness_weighted_correct_rate']
    team_correct_rate_mean /= 4
    team_hardness_weighted_correct_rate_mean /= 4
    for member in individual_performance.keys():
        member_correct_rate = individual_performance[member]['correct_rate']
        member_hardness_weighted_correct_rate = individual_performance[member]['hardness_weighted_correct_rate']
        individual_performance[member]['correct_rate_minus_average_in_team'] = (
            member_correct_rate - team_correct_rate_mean)
        individual_performance[member]['hardness_weighted_correct_rate_minus_average_in_team'] = (
            member_hardness_weighted_correct_rate - team_hardness_weighted_correct_rate_mean)
```

```python
len(individual_performance_rates)
```

```python
amount_of_appraise_by_others = defaultdict(lambda: defaultdict(list))
amount_of_self_appraise = defaultdict(lambda: defaultdict(list))
value_in_left_eigvec_full_appraisals = defaultdict(lambda: defaultdict(list))
value_in_left_eigvec_relativereflected_appraisals = defaultdict(lambda: defaultdict(list))

for team_id, team_log in data.items():
    if team_id in individual_performance_rates:
        if len(team_log.member_influences) < 1 or np.sum(team_log.member_influences_from_data[-1]) != 16:
            continue
        for member_influence in team_log.member_influences:
            influence_matrix = utils.make_matrix_row_stochastic(
                np.matrix(member_influence) / 100)
            relative_reflected_appraisal_matrix = utils.get_relative_reflected_appraisal_matrix(influence_matrix)
            full_appraisals_in_this_team = utils.get_stationary_distribution(
                influence_matrix,
                aperiodic_irreducible_eps=0)
            skip = False
            for i in range(4):
                if np.sum(relative_reflected_appraisal_matrix[i, :]) == 0:
                    print('Due to self appraisal equals to one, team {} was skipped.'.format(
                        team_id))
                    skip = True
                    break
            if not skip:
                relative_reflected_appraisals_in_this_team = utils.get_stationary_distribution(
                    relative_reflected_appraisal_matrix,
                    aperiodic_irreducible_eps=0)

            for index, member in enumerate(team_log.members):
                if member in individual_performance_rates[team_id]:
                    amount_of_appraise_by_others[team_id][member].append(
                        utils.get_average_influence_on_others(
                            influence_matrix, index))
                    amount_of_self_appraise[team_id][member].append(influence_matrix[index, index])
                    value_in_left_eigvec_full_appraisals[team_id][member].append(
                        full_appraisals_in_this_team[index])
                    if not skip:
                        value_in_left_eigvec_relativereflected_appraisals[team_id][member].append(
                            relative_reflected_appraisals_in_this_team[index])
```

```python
len(amount_of_appraise_by_others)
```

```python
maxlag = 2

all_tests = 0
ignored = 0
sig_rvals = []
sig_causes = []

for team_id, self_appraises in amount_of_self_appraise.items():
    for member, member_self_appraise in self_appraises.items():
        member_appraise_by_others = amount_of_appraise_by_others[team_id][member]
        perf = individual_performance_rates[team_id][member]
        (r, p) = pearsonr(member_self_appraise, member_appraise_by_others)
        all_tests += 1
        if p < 0.05:
            sig_rvals.append([team_id, member, r])
        try:
            causality_res = grangercausalitytests(
                np.column_stack((member_self_appraise, member_appraise_by_others)),
                maxlag=maxlag,
                verbose=False)
            found = False
            for lag in range(1, maxlag+1):
                res = causality_res[lag][0]
                for test_values in res.values():
                    if test_values[1] < 0.05:
                        sig_causes.append([team_id, member, res, perf])
                        found = True
                        break
                if found:
                    break
        except ValueError:
            try:
                causality_res = grangercausalitytests(
                    np.column_stack((member_self_appraise, member_appraise_by_others)),
                    maxlag=maxlag,
                    verbose=True)
            except:
                pass
            ignored += 1
            print('ignored.')
```
